```python
url = 'https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'
fuel_data = pd.read_csv(url, error_bad_lines=False)
fuel_data.describe(include='all')
```

```python
fuel_df1 = fuel_data.iloc[0:19000].reset_index(drop=True)
fuel_df2 = fuel_data.iloc[19000:].reset_index(drop=True)
```

```python
unit_df = fuel_data.groupby('fuel_unit')['fuel_unit'].count().to_frame(name='count')
unit_df['count']
```

```python
#importing library
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#Setting figure size
plt.figure(figsize=(7,4))

g = sns.barplot(data=unit_df,x=unit_df.index,y=unit_df['count'])
plt.xlabel('Fuel Unit')
#because of the extreme y-range, set log scale
g.set_yscale("log")
g.set_ylim(1,12000)
plt.xticks(rotation=30)

```

```python
sample_df = fuel_data.sample(n=50, random_state=4)
sns.regplot(x=sample_df["utility_id_ferc1"], y=sample_df["fuel_cost_per_mmbtu"],
fit_reg=False)
```
